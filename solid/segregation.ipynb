Interfaz para el huésped

```python
from abc import ABC, abstractmethod

class GuestReservationSystem(ABC):
    @abstractmethod
    def search_available_rooms(self, check_in_date, check_out_date):
        pass

    @abstractmethod
    def make_reservation(self, room_type, check_in_date, check_out_date, guest_info):
        pass

    @abstractmethod
    def cancel_reservation(self, reservation_id):
        pass
```
